```python

days = int(input("Enter number of days: "))

total_sales = 0
for i in range(1, days + 1):
    sales = float(input(f"Enter sales for day {i}: "))
    total_sales += sales
average_sales = total_sales / days

print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
```
