```python
n = int(input("Enter number of interns: "))
present_count = 0
absent_count = 0
for i in range(1, n + 1):
    status = input(f"Enter attendance for intern {i} (P/A): ")
    if status == "P" or status == "p":
        present_count += 1
    elif status == "A" or status == "a":
        absent_count += 1
    else:
        print("Invalid input! Please enter P or A.")
print("\nAttendance Summary")
print("Total Present:", present_count)
print("Total Absent:", absent_count)
```
