```python
distance = int(input("Enter delivery distance in km: "))
if distance <= 5:
    charge = 0
elif distance <= 15:
    charge = (distance - 5) * 10
else:
    charge = (10 * 10) + (distance - 15) * 20
print("Total Delivery Charge: ", charge)
```
