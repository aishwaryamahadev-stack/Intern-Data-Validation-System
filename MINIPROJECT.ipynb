```python
def login(username, password):
    if username == "dd_admin" and password == "dd@123":
        print("Login Successful\n")
        return True
    else:
        print("Access Denied")
        return False
user = input("Enter Username: ")
pwd = input("Enter Password: ")
if not login(user, pwd):
    exit()
def register_employee():
    emp_id = input("Enter Employee ID: ")
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    department = input("Enter Department: ")
    basic_salary = float(input("Enter Basic Salary: "))

    if age < 18:
        print("Error: Age must be 18 or above")
        return None
    elif basic_salary <= 0:
        print("Error: Salary must be greater than 0")
        return None
    else:
        print("Employee Registered Successfully\n")
        return emp_id, name, age, department, basic_salary


employee = register_employee()
if employee is None:
    exit()
present_days = 0

print("Enter Attendance for 5 Days (P/A):")
for i in range(1, 6):
    status = input(f"Day {i}: ").upper()
    if status == "P":
        present_days += 1

print("Total Present Days:", present_days)

if present_days >= 4:
    attendance_status = "Regular"
elif present_days == 3:
    attendance_status = "Warning"
else:
    attendance_status = "Irregular"

print("Attendance Status:", attendance_status, "\n")
def calculate_salary(basic_salary, present_days):
    hra = basic_salary * 0.20
    da = basic_salary * 0.10
    pf = basic_salary * 0.12

    if present_days < 3:
        hra = 0
        da = 0

    net_salary = basic_salary + hra + da - pf
    return net_salary


net_salary = calculate_salary(employee[4], present_days)
print("Net Salary:", net_salary, "\n")
total_rating = 0

print("Enter 3 Performance Ratings (1 to 5):")
for i in range(1, 4):
    rating = int(input(f"Rating {i}: "))
    total_rating += rating

average = total_rating / 3
print("Average Rating:", average)

if average >= 4:
    performance = "Excellent"
elif average >= 3:
    performance = "Good"
else:
    performance = "Needs Improvement"

print("Performance Status:", performance)
```
