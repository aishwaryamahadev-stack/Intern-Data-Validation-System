```python
def calculate_discount(price,customer_type):

    customer_type = customer_type.lower()

    if customer_type == "regular":
        discount = 0.05
    elif customer_type == "premium":
        discount = 0.15
    elif customer_type == "employee":
        discount = 0.25
    final_price = price * (1 - discount)
    return final_price
price = 1000
for customer in ["Regular", "Premium", "Employee"]:
    discounted_price = calculate_discount(price, customer)
    print(f"{customer} customer: Final price = â‚¹{discounted_price:.2f}")
```
