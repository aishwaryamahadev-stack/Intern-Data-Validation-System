```python
def check_attendance(login_time):
    if login_time < "09:30":
        return "Present"
    elif "09:30" <= login_time <= "10:00":
        return "Late"
    else:
        return "Absent"
print(check_attendance("09:15"))  # Present
print(check_attendance("09:45"))  # Late
print(check_attendance("10:10"))  # Absent
```
