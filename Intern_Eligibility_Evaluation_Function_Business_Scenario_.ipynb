```python
def check_intern_eligibility(age, percentage):
    if age >= 18 and percentage >= 60:
        return "Eligible: Age and graduation percentage meet the requirements."
    elif age < 18 and percentage < 60:
        return "Not Eligible: Age is below 18 and graduation percentage is below 60%."
    elif age < 18:
        return "Not Eligible: Age must be 18 or above."
    else:
        return "Not Eligible: Graduation percentage must be 60% or above."
age = int(input("Enter age: "))
percentage = float(input("Enter graduation percentage: "))
result = check_intern_eligibility(age, percentage)
print(result)
18
```
